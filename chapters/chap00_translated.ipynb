# 序文

## この本は誰のためのものですか？

プログラミングを学びたいと思っている方には、ぴったりの本です。Pythonは初心者にとって最適なプログラミング言語の一つであり、最も需要が高いスキルの一つでもあります。

また、今が学ぶのに絶好のタイミングです。現在は、プログラミングの学習がこれまでになく簡単になっています。ChatGPTのようなバーチャルアシスタントを活用することで、一人で学ぶ必要はありません。本書では、これらのツールを使って学習を加速する方法を提案します。

この本は主に、プログラムを書いたことがない方や、他のプログラミング言語に多少の経験がある方を対象としています。Pythonに十分な経験がある方にとっては、最初の数章は退屈に感じられるかもしれません。

プログラミングを学ぶ際の課題の一つは、*二つ*の言語を学ばなければならないことです。一つはプログラミング言語そのもので、もう一つはプログラムについて話すための語彙です。プログラミング言語だけを学んだ場合、エラーメッセージの解釈やドキュメントの読解、他の人との会話、バーチャルアシスタントの利用の際に問題が生じる可能性があります。もし、少しプログラミングを経験したことがあっても、この第二の言語を学んでいない場合は、この本が役立つことを願っています。

## 本書の目標

この本を書くにあたって、私は語彙の選定に注意を払いました。
各用語は初めて登場した際に定義しています。
各章の終わりには、その章で紹介した用語を復習する用語集もあります。

また、私は簡潔であることを心掛けました。
読者が本を読むのに必要とする精神的な労力を減らすことで、プログラミングに割く余力を増やせるからです。

しかし、プログラミングを学ぶためには本を読むだけでは不十分です――練習が必要です。
そのため、この本には各章の終わりに、学んだことを練習できる演習問題が含まれています。

注意深く読み、演習問題に一貫して取り組めば、進歩が見られるでしょう。
しかし、今のうちに警告しておきます――プログラミングを学ぶことは簡単ではなく、経験豊富なプログラマーでさえ挫折することがあります。
進むにつれて、正しいプログラムを書くための戦略や、誤ったプログラムを修正する方法を提案します。

## 本書のナビゲーション

本書の各章は前の章に基づいているため、順番に読み進め、先へ進む前に演習に取り組むことをお勧めします。

最初の6つの章では、算術、条件分岐、ループといった基本的な要素を紹介します。また、プログラミングにおいて最も重要な概念である関数と、それを活用する強力な方法である再帰も紹介します。

第7章と第8章では、文字や単語、文章を表現できる文字列を紹介し、それらを扱うアルゴリズムについて説明します。

第9章から第12章では、リスト、辞書、タプルといったPythonのコアデータ構造を紹介します。これらは効率的なプログラムを書くための強力なツールです。第12章では、テキストを分析し、新しいテキストをランダムに生成するアルゴリズムを紹介します。このようなアルゴリズムは、大規模言語モデル（LLM）の中核を成しており、この章ではChatGPTのようなツールがどのように機能するかのアイデアを提供します。

第13章は、長期的なデータ保存方法であるファイルとデータベースについてです。演習として、ファイルシステムを検索して重複ファイルを見つけるプログラムを書くことができます。

第14章から第17章では、オブジェクト指向プログラミング（OOP）について紹介します。OOPはプログラムとデータを整理する方法です。多くのPythonライブラリはオブジェクト指向スタイルで書かれているため、これらの章はライブラリの設計を理解し、自分のオブジェクトを定義するのに役立ちます。

本書の目的は、Python言語全体を網羅することではありません。むしろ、最小限の概念で最大の能力を提供する言語のサブセットに焦点を当てています。それでもなお、Pythonには一般的な問題を効率的に解決するための多くの機能があります。第18章ではこれらの機能の一部を紹介します。

最後に、第19章では別れの言葉とプログラミングの旅を続けるための提案をお伝えします。

## 第3版の新しい点は？

この版での最大の変更点は、Jupyterノートブックと仮想アシスタントという2つの新しい技術によってもたらされました。

この本の各章はJupyterノートブックとして提供されており、通常のテキストとコードの両方が含まれています。
私にとってはコードを書き、そのテストを行い、テキストと一貫性を保つことが容易になります。
読者にとっては、コードを実行したり、修正したり、演習に取り組んだりすることが、一つの場所で行えるようになります。
ノートブックを使用するための手順は第1章にあります。

もう一つの大きな変更点は、ChatGPTのような仮想アシスタントを活用して学習を加速するためのアドバイスを追加したことです。
前版が出版された2016年当時、これらのツールの前身はあまり実用的ではなく、大多数の人はそれらの存在を知らなかったでしょう。
現在では、ソフトウェア開発の標準的なツールとなっており、プログラミングを学ぶ上で、さらには他の多くのことを学ぶ上でも、非常に重要な変革手段になると考えています。

他の変更点は、第2版に対する私の後悔から動機づけられました。

最初の後悔は、ソフトウェアテストを強調しなかったことです。
2016年時点ですでに残念な欠落でしたが、現在は仮想アシスタントの登場により、テストの自動化がさらに重要になっています。
そこでこの版では、Pythonで最も広く使われているテストツールである`doctest`と`unittest`を取り上げ、いくつかの演習を通じてそれらを実践できるようにしました。

もう一つの後悔は、第2版の演習が不均一であったことです。
いくつかは非常に興味深いものである一方、いくつかは難しすぎるものでした。
Jupyterノートブックに移行したおかげで、より魅力的で効果的な演習の流れを開発し、テストすることができました。

この改訂では、トピックの順番はほとんど同じですが、いくつかの章を再配置し、2つの短い章を1つにまとめました。
また、正規表現を含むように文字列に関するカバレッジを拡大しました。

いくつかの章ではタートルグラフィックスを使用しています。
以前の版ではPythonの`turtle`モジュールを使用していましたが、Jupyterノートブックでは動作しないため、新しいタートルモジュールに置き換え、それを使いやすくしました。

最後に、文章の多くを書き直し、必要な箇所を明確化し、あまり簡潔でない部分を削減しました。

私はこの新しい版をとても誇りに思っています。気に入っていただけることを願っています！

## はじめに

Pythonを含むほとんどのプログラミング言語において、プログラムを書くための多くのツールがあります。これらのツールは統合開発環境（IDE）と呼ばれます。一般的に、IDEには2つの種類があります。

* あるものはコードを含むファイルで作業を行い、これらのファイルを編集し実行するためのツールを提供します。

* 他は主に文章とコードを含むノートブックで作業を行います。

初心者には、Jupyterのようなノートブック開発環境から始めることをお勧めします。

この本のためのノートブックは、オンラインリポジトリ<https://allendowney.github.io/ThinkPython>から入手可能です。

これらを使用する方法は2つあります：

* ノートブックをダウンロードし、自分のコンピュータで実行することができます。その場合、PythonとJupyterをインストールする必要がありますが、それほど難しくはありません。しかし、Pythonを学びたいだけなら、ソフトウェアのインストールに多くの時間をかけるのはストレスがたまることがあります。

* もう1つの方法は、Colabでノートブックを実行することです。Colabはウェブブラウザで動作するJupyter環境なので、何もインストールする必要がありません。ColabはGoogleが運営しており、無料で使用できます。

もしあなたがこれから始めるのであれば、Colabから始めることを強くお勧めします。
